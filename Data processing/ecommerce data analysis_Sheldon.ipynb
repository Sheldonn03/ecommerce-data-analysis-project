```python
df_main = df[
    (df['Membership_Years'] <= df['Membership_Years'].quantile(0.99)) &
    (df['Total_Purchases'] <= df['Total_Purchases'].quantile(0.99)) &
    (df['Lifetime_Value'] <= df['Lifetime_Value'].quantile(0.99))
].copy()

df_main['Membership_Group'] = pd.cut(
    df_main['Membership_Years'],
    bins=[0, 1, 3, 5, df_main['Membership_Years'].max()],
    labels=['0-1 Years', '1-3 Years', '3-5 Years', '5+ Years'],
    include_lowest=True
)

df_main['Purchase_Group'] = pd.cut(
    df_main['Total_Purchases'],
    bins=[0, 8, 15, 25, df_main['Total_Purchases'].max()],
    labels=['Low Purchases', 'Medium Purchases', 'High Purchases', 'Very High Purchases'],
    include_lowest=True
)

grouped = (
    df_main.groupby(['Membership_Group', 'Purchase_Group'], observed=True)
    .agg(avg_lifetime_value=('Lifetime_Value', 'mean'))
    .reset_index()
)

plt.figure(figsize=(12, 6))

ax = sns.barplot(
    data=grouped,
    x='Membership_Group',
    y='avg_lifetime_value',
    hue='Purchase_Group',
    palette='Blues'
)

for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.0f',
        padding=3,
        fontsize=8
    )

plt.title('Average Lifetime Value by Membership Years and Purchase Frequency')
plt.xlabel('Membership Duration')
plt.ylabel('Average Lifetime Value')
plt.legend(title='Purchase Group', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

Sentiment Analysis

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

df['sentiment_score'] = df['Reviews'].apply(
    lambda x: sia.polarity_scores(str(x))['compound']
)
```

```python
def classify_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_label'] = df['sentiment_score'].apply(classify_sentiment)

df[['Reviews', 'sentiment_score', 'sentiment_label']].head()
```

```python
sentiment_counts = df['sentiment_label'].value_counts()

sentiment_order = ['Positive', 'Neutral', 'Negative']
sentiment_counts = sentiment_counts.reindex(sentiment_order)

colors = ['#4E79A7', '#F28E2B', '#E15759']

plt.figure(figsize=(8, 6))

ax = sns.barplot(
    x=sentiment_counts.index,
    y=sentiment_counts.values,
    palette=colors
)

# 手動在 bar 上方加數字
for i, value in enumerate(sentiment_counts.values):
    ax.text(
        i,
        value,
        f'{value:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.title('Sentiment Distribution of Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.ylim(0, sentiment_counts.max() * 1.15)

plt.show()
```

The sentiment analysis classifies product reviews into positive, neutral, and negative groups using VADER compound sentiment scores.

Positive reviews generally indicate smoother shopping experiences, stronger product satisfaction, and higher likelihood of repeat purchase.
Negative reviews often mention service issues, return concerns, checkout friction, or weak perceived value.
